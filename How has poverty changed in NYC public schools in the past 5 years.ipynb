```python
#import Python packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import pyplot
```

```python
#read .csv file, first take a look at 2015-2016 school year data for schools with greater than 500 students
nyckids = pd.read_csv('NYC Educational Data.csv',delimiter=',')
nyc2015 = nyckids.loc[nyckids['Year'] == '2015-16']
nyc2015 = nyc2015.loc[nyckids['Total Enrollment'] > '500']
nyc2015 = nyc2015.set_index('Year')
nyc2015.head(10)
```

```python
#convert '%' sign to a string 
nyc2015['% Poverty'] = nyc2015['% Poverty'].str.replace('%','').astype(float)
nyc2015['% White'] = nyc2015['% White'].str.replace('%','').astype(float)
nyc2015.head(5)
```

```python
#sort all the schools by % Poverty
sorted = nyc2015.sort_values(by=['% Poverty'])
sorted.head(5)
```

```python
#Sort schools by % Poverty, all schools below 20%
bottom20 = sorted.loc[sorted['% Poverty'] < 20.0]
bottom20.head(5)
```

```python
#All schools with above 80% poverty, ascending order
top20 = sorted.loc[sorted['% Poverty'] > 80.0]
top20.head(5)
```

```python
#mean percentage of students identified as White in the bottom 20% and top 20% schools
meanpctwhite_bottom = bottom20['% White'].mean()
meanpctwhite_top = top20['% White'].mean()
print (meanpctwhite_bottom)
print (meanpctwhite_top)
```

```python
#replacing '%' sign into string and commas for the entire table
nyckids['% Poverty'] = nyckids['% Poverty'].str.replace('%','').astype(float)
nyckids['# Poverty'] = nyckids['# Poverty'].str.replace(',','').astype(float)
nyckids['% White'] = nyckids['% White'].str.replace('%','').astype(float)
nyckids['# White'] = nyckids['# White'].str.replace(',','').astype(float)
nyckids['% English Language Learners'] = nyckids['% English Language Learners'].str.replace('%','').astype(float)
nyckids['# English Language Learners'] = nyckids['# English Language Learners'].str.replace(',','').astype(float)
nyckids = nyckids[['DBN','Year','School Name','# White','% White','# Poverty','% Poverty','% English Language Learners','# English Language Learners']]
nyckids.head(10)
```

```python
#in order to calculate the change in poverty throughout time, have to first create empty Pandas df with the school name and change in poverty cols
#then empty arrays with the % Pov, Pov Num, ELL Num, etc in order to fill fill them with the values for each year, per school
name = 'P.S. 015 Roberto Clemente'
final_table = pd.DataFrame(columns = ['School Name','% Poverty Delta'])
l = []
pov_num = []
ell_num = []
ell_pct = []
white_num = []
white_pct = []

#for each school
for index, row in nyckids.iterrows():
    #if the school name is the same, append values for each column header
    if name == row['School Name']:
        l.append(row['% Poverty'])
        pov_num.append(row['# Poverty'])
        ell_num.append(row['# English Language Learners'])
        ell_pct.append(row['% English Language Learners'])
        white_num.append(row['# White'])
        white_pct.append(row['% White'])
    else:
        #if there's a new school name, calculate the delta of the array by subtracting the last value with the first, append to the final table
        delta = l[-1] - l[0]
        pov_delta = pov_num[-1] - pov_num[0]
        ellnum_delta = ell_num[-1] - ell_num[0]
        ellpct_delta = ell_pct[-1] - ell_num[0]
        whitenum_delta = white_num[-1] - white_num[0]
        whitepct_delta = white_pct[-1] - white_pct[0]
        final_table = final_table.append({'School Name':nyckids.loc[index-1,'School Name'],'% White Delta':whitepct_delta,'# White Delta':whitenum_delta,'% Poverty Delta':delta,'# Poverty Delta':pov_delta,'% ELL Delta':ellpct_delta,'# ELL Delta':ellnum_delta}, ignore_index=True)
        l = []
        pov_num = []
        ell_num = []
        ell_pct = []
        white_num = []
        white_pct = []
        l.append(row['% Poverty'])
        pov_num.append(row['# Poverty'])
        ell_num.append(row['# English Language Learners'])
        ell_pct.append(row['% English Language Learners'])
        white_num.append(row['# White'])
        white_pct.append(row['% White'])
        name = row['School Name']
            
final_table.head(10) 

#nyckids_delta  = pd.concat([a,b],axis=1).dropna()
#nyckids_delta
```
